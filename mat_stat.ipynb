# Методы математической статистики

## Основы прикладной статистики

$\blacktriangleright$ Основные понятия математической статистики 

Математическая статистика -- наука, изучающая закономерности в явлениях, которые носят случайный характер, но являются массовыми. При этом данные закономерности исследуются на основе конечного числа наблюдений.

Методы математической статистики -- это методы теории вероятностей, математического анализа и дифференциальных уравнений, специальные методы, используемые для исследования массовых совокупностей.

Все методы математической статистики можно подразделить на параметрические методы, основанные на использовании знаний о вероятностной модели, и непараметрические методы, когда априорных представлений о виде модели нет или она не используется.  

Основными задачами математической статистики являются:
1. Оценка неизвестной функции распределения и функции плотности вероятности.
2. Проверка статистических гипотез.
3. Оценка неизвестных параметров распределения.
4. Исследование статистических зависимостей.

Генеральной совокупностью называется вся мыслимая совокупность исследуемых объектов. 

Выборочной совокупностью называется совокупность объектов, отобранная для непосредственного изучения. Выборочную совокупность также называют выборкой. 

Объемом совокупности (генеральной или выборочной) называется число ее объектов.

При этом различают:
1. Метод сплошных наблюдений: исследуется вся генеральная совокупность. Однако на практике данный метод, как правило, требует больших затрат и является достаточно трудоемким, поэтому чаще прибегают к выборочному методу.
2. Выборочный метод исследуется только выборочная совокупность и на основании этого выборочного обследования делается вывод о всей генеральной совокупности.

Схема статистического исследования включает два основных этапа:
1. Выделяется совокупность объектов и исследуемый признак.
2. Исследуемый признак выражается числовой величиной.

Пусть посредством независимых испытаний, проводимых в одинаковых условиях, получены некоторые значения исследуемого признака. Расположим эти значения в порядке возрастания $x_1 \leqslant x_2 \leqslant  ... \leqslant x_n$.

Последовательность наблюдаемых значений $x_1, x_2, \ldots, x_n$, записанных в возрастающем порядке, называется дискретным вариационным рядом, а сами значения -- вариантами. 

Значение $x_i$ иногда называют порядковой  статистикой. Индекс $i$ указывает на порядковый номер элемента в вариационном ряду. 

Выборочный $\alpha$-квантиль или выборочный квантиль порядка $\alpha$, $\alpha \in (0,\,1)$, есть статистика, равная элементу вариационного ряда с номером $[m\alpha+1]$ (целая часть от $m\alpha+1$).

Частотой $n_i$ называется число, показывающее, сколько раз встречается вариант $x_i$. 

Относительной частотой $\omega_i$ называется отношение соответствующей частоты к общему объему выборки, то есть 
$$
\omega_i = \dfrac{n_i}{n}.
$$ 

Пусть дан некоторый вариационный ряд $x_1, x_2, \ldots, x_n$ и соответствующие частоты $n_1, n_2, \ldots, n_n$ вариант. 

Выборочной средней называется величина 
$$
\bar{x} = \dfrac{n_1\cdot x_1+ n_2\cdot x_2 + \ldots + n_n\cdot x_n}{n_1+n_2+\ldots + n_n}
$$

Выборочной дисперсией называется величина 
$$
\bar{D} = \dfrac{( x_1 - \bar{x})^2n_1+ ( x_2 - \bar{x})^2n_2 + \ldots + (x_n - \bar{x})^2n_n}{n_1+n_2+\ldots + n_n}
$$

Выборочным средним квадратическим отклонением  называется арифметический квадратный корень из дисперсии.

Медианой вариационного ряда называется значение признака, приходящееся на середину ранжированного ряда наблюдений.  

Модой вариационного ряда называется вариант, которому соответствует наибольшая частота.

Существуют также и другие выборочные статистики, например, коэффициент 
вариации, коэффициент Лексиса, асимметрия. Сложность в их использовании заключается в интерпретации.


Прикладная статистика -- наука о методах обработки статистических данных.

По типу решаемых задач прикладная статистика делится на разделы:
- описание данных;
- оценивание;
- проверка гипотез.

В прикладной статистике для описания данных часто используются случайные величины.

Под **случайной величиной** понимается величина, которая в результате опыта со случайным исходом принимает то или иное значение. 

Возможные значения случайной величины образуют множество $\Theta$, которое принято называть множеством возможных значений случайной величины.

Примеры случайных величин: доход клиента банка.

Замечание: Случайные величины принято обозначать большими буквами, а их возможные значения -- маленькими. Например, случайная величина $X$ имеет значения $x_1, x_2, \ldots, x_n$.

В зависимости от множества значений $\Theta$ принято выделять следующие типы случайных величин:
- дискретные --- если множество значений $\Theta$ счетно или конечно (элементы $\Theta$ можно перенумеровать);
- непрерывные --- если множество значений $\Theta$ несчетно. 

Примеры: число очков на грани игральной кости -- дискретная случайная величина; скорость, доход --- непрерывная случайная величина.

Случайная величина описывается законом распределения.

Законом распределения случайной величины называется любое правило 
(таблица, функция), позволяющее находить вероятности всевозможных событий, связанных со случайной величиной (например, вероятность того, что она примет какое-то значение или попадет на какой-то интервал). 

Наиболее простой способ задания закона распределения -- табличный.
$$
X: \begin{array}{|c|c|c|c|}
x_1 & x_2 & \ldots & x_n \\
\hline
p_1 & p_2 & \ldots & p_n
\end{array}
$$

Наиболее общей формой закона распределения, пригодной для всех случайных величин (как дискретных, так и непрерывных), является функция распределения. 

Функцией распределения случайной величины $X$ называется вероятность того, что она примет значение меньшее, чем заданное $x$:
$$
F(x) = P(X<x).
$$

Свойства функции распределения:

1) $F(x)$ -- неубывающая функция своего аргумента, т. е. при $x_2 > x_1$ $F(x_2)\geqslant F(x_1)$.

2) $F(-\infty) = 0. $ 

3) $F(+\infty) = 1. $

Замечание: Случайная величина $X$ является непрерывной, если ее функция распределения не только непрерывна в любой точке, но и дифференцируема всюду, 
кроме, может быть, отдельных точек, где она терпит излом.

Плотностью распределения (или плотностью вероятности иногда просто плотностью) непрерывной случайной величины $X$ в точке $x$ называется производная ее 
функции распределения в этой точке.
$$
f(x) = F'(x) = \dfrac{\partial}{\partial x} F(x).
$$

Плотность распределения $f(x)$, как и функция распределения $F(x)$, является одной из форм закона распределения; в отличие от функции распределения, эта 
форма не универсальна: она существует только для непрерывных случайных величин.

График плотности распределения $f(x)$ называется кривой распределения.

Числовые параметры случайной величины, характеризующие существенные черты распределения, принято называть числовыми характеристиками случайной величины. 

Среди числовых характеристик случайных величин выделяют: параметры положения, масштабные параметры и параметры формы.

Параметр положения $\gamma$ определяет положение области значений распределения на оси абсцисс (оси х). Параметрами положения являются: математическое ожидание, мода и медиана.

Математическое ожидание случайной величины определяется следующим образом:
- для дискретных с.в. 
$$
E(X) = \sum\limits_{i=1}^{n}x_ip_i
$$ 
- для непрерывных с.в. 
$$
E(X) = \int\limits_{-\infty}^{+\infty}xf(x)\,dx
$$ 

Масштабный параметр $\beta$ определяет масштаб (или единицы) измерения значений в диапазоне распределения. 

Параметр формы $\alpha$ в отличие от параметра положения и масштабного  
параметра определяет основную форму распределения в общем семействе  
распределений.

## Корреляционный анализ

### Определение корреляционного анализа [1]

Пусть имеется некоторая выборочная совокупность объектов, каждый из которых описывается единым набором признаков $x_1, x_2, \ldots, x_n, y$. Представим эти данные в виде матрицы <<объект-свойство>>

$$
\begin{array}{ccccc}
x_{11} & x_{12} & \ldots  & x_{1n} & y_1\\
x_{21} & x_{22} & \ldots  & x_{2n} & y_2\\
\vdots  & \vdots & \ddots  & \vdots & \vdots\\
x_{m1} & x_{m2} & \ldots  & x_{mn} & y_m
\end{array}
$$  

Одна из центральных проблем прикладной статистики состоит в статистическом исследовании зависимостей, а именно в выявлении и описании парных и множественных статистических связей, существующих между признаками исследуемых объектов. Иными словами, для анализируемой выборки необходимо, прежде всего, выяснить имеется ли статистическая зависимость между признаками объектов или они являются статистически независимыми. 

Одним из методов проведения статистического исследования зависимостей является корреляционный анализ. 

Корреляционным анализом называется совокупность методов статистического анализа, позволяющих:
- выбрать (с учетом специфики и природы анализируемых переменных) подходящий измеритель статистической связи (коэффициент корреляции, корреляционное отношение, какую-либо информационную характеристику связи, ранговый коэффициент корреляции и т. п.);
- оценить (с помощью точечной и интервальной оценок) его числовое значение по имеющимся выборочным данным;
- проверить гипотезу о том, что полученное числовое  значение анализируемого измерителя связи действительно свидетельствует о наличии статистической связи;
- определить структуру связей между компонентами исследуемого многомерного признака, сопоставив каждой паре компонент двоичный ответ («связь есть» или «связи нет»).

### Статистическая зависимость

Замечание (о понятии статистической зависимости). Две случайные величины могут  быть связаны либо функциональной зависимостью, либо статистической зависимостью, либо быть независимыми. 

Функциональная зависимость между величинами означает, что каждому значению одной случайной величины поставлено в соответствие единственное значение другой случайной величины. Функциональные зависимости между случайными величинами встречаются редко, поскольку эти случайные величины или одна из них подвержены еще воздействию других случайных величин.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#генерируем две нормально распределенные случайные величины случайные величины 
x = np.random.normal(2, 4, 100)
z = np.random.normal(0, 1, 100)
y = 2*x + 3 + z
f = 2*x + 3
```

```python
fig, ax = plt.subplots(1,3, figsize = (20,4))
plt.subplot(131)
plt.scatter(x,z, 
            color = 'red',
            edgecolors = 'black')
plt.xlabel('Случайная величина X')
plt.ylabel('Случайная величина Z')
plt.title('Диаграмма рассеивания -- независимые величины')

plt.subplot(132)
plt.scatter(x,y,
           color = 'green',
           edgecolors = 'black')
plt.xlabel('Случайная величина X')
plt.ylabel('Случайная величина Y')
plt.title('Диаграмма рассеивания -- стастистическая зависимость')

plt.subplot(133)
plt.scatter(x,f,
           color = 'blue',
           edgecolors = 'black')
plt.xlabel('Случайная величина X')
plt.ylabel('Случайная величина Y')
plt.title('Диаграмма рассеивания -- функциональная зависимость')
plt.show()
```

Статистической называют зависимость, при которой изменение одной из величин влечет изменение распределения другой. 

Статистическая зависимость, при которой изменение одной величины приводит к изменению значения выборочного среднего другой величины, называется корреляционной.

Особенности корреляционной зависимости:
- наличие корреляционной зависимости между переменными не всегда означает наличие непосредственной связи этих величин друг с другом;
- на основании корреляции можно утверждать лишь о степени связи между переменными, но не о существовании причинно-следственной зависимости между ними.

Корреляционные зависимости различаются по тесноте (силе) связи и количеству признаков. 

По тесноте корреляционной связи принято выделять: 1) функциональную, 2) тесную (сильную), 3) среднюю (умеренную), 4) слабую и 5) нулевую (отсутствующую) виды связи. 

По количеству признаков корреляция может быть парной (между двумя признаками) и множественной (между несколькими признаками).
Форма парной корреляции может быть линейной, описываемой линейной функцией регрессии, и нелинейной, описываемой нелинейными функциями.

Парная линейная корреляция, в свою очередь, может быть положительной («прямой») и отрицательной («обратной»). При положительной корреляции при возрастании одного признака в среднем увеличивается другой, в случае же отрицательной корреляции при возрастании одного признака другой в среднем уменьшается.

Для оценки тесноты линейной связи между случайными величинами $X$ и $Y$ служит коэффициент корреляции, определяемый по следующей формуле
$$
r_{xy} = \dfrac{E(XY)-E(X)E(Y)}{\sigma_X \sigma_Y},
$$
где $E(\cdot)$ -- оператор математического ожидания; $\sigma_X$ и $\sigma_Y$ -- среднеквадратические отклонения случайных величин $X$ и $Y$.

В литературе данный коэффициент называется коэффициентом линейной корреляции Пирсона. 

Выборочный коэффициент корреляции для величин $X$ и $Y$ вычисляется по формуле:
$$
\hat{r_{xy}} = \dfrac{\sum\limits_{i=1}^{m}(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum\limits_{i=1}^{m}(x_i-\bar{x})^2(y_i-\bar{y})^2}},
$$

Коэффициент корреляции показывает, насколько связь между случайными величинами близка к строго линейной. 

Свойства коэффициента корреляции:
1) коэффициент корреляции симметричен относительно переменных $X$ и $Y$: $r_{xy}=r_{yx}$;

2) коэффициент корреляции есть безразмерная величина;

3) коэффициент корреляции принимает значения на отрезке $[-1;1]$: $-1 \leqslant r_{xy} \leqslant 1 $;

4) если $|r_{xy}| \to 1$, то связь между $Y$ и $X$ становится все более тесной и при $|r_{xy}| = 1$ корреляционная связь превращается в функциональную;

5) если $|r_{xy}| \to 0$, то корреляционная связь между величинами $X$ и $Y$ ослабевает, а при $|r_{xy}| = 0$ связь между этими переменными отсутствует: $X$ и $Y$ ведут себя как независимые случайные величины.

Направление линейной корреляционной связи определяется знаком коэффициента корреляции: для «прямой», положительной связи $r_{xy} > 0$, для «обратной», отрицательной связи $r_{xy} < 0$.

Теснота линейной связи между случайными величинами определяется абсолютной величиной коэффициента корреляции:

|Значение модуля коэффициента корреляции|Теснота связи|
|--:|:--|
|$|r_{xy}|=1$|величины связаны линейной функциональной зависимостью|
|$0.95 \leqslant |r_{xy}| < 1$| связь очень сильная, практически функциональная|
|$0.75 \leqslant |r_{xy}| < 0.95$ | связь тесная (сильная)|
|$0.5 \leqslant |r_{xy}| < 0.75$ | связь средняя (умеренная)|
|$0.2 \leqslant |r_{xy}| < 0.5$ | связь слабая|
|$0 \leqslant |r_{xy}| < 0.2$ | практически нет связи|

**Ранговая корреляция**

Для определения тесноты связи между признаками, измеренными в порядковых (ранговых) шкалах, применяются методы ранговой корреляции. К ним относятся: коэффициенты ранговой корреляции Спирмена и Кендалла.

**Коэффициент ранговой корреляции Спирмена**

Пусть $R^X$ и $R^Y$ последовательности рангов (значений) случайных величин $X$ и $Y$, выраженных в порядковых шкалах. Коэффициент корреляции Спирмена определяется формулой:
$$
\rho_{XY}^{S} = 1 - \dfrac{6\sum\limits_{i=1}^{m}(R_i^X-R_i^Y)^2}{m(m^2-1)}. 
$$
Его значения находятся в интервале от $-1$ до $+1$.

**Коэффициент ранговой корреляции Кендалла**

Пусть $R^X$ и $R^Y$ последовательности рангов (значений) случайных величин $X$ и $Y$, выраженных в порядковых шкалах. Упорядочим значения $R^X$ по возрастанию. Пару значений $R_i^Y$ и $R_j^Y$ ($i<j$) называют инверсией, если в последовательности $R^Y$ имеет место $R_i^X > R_j^Y$. Обозначим через $Q$ множество таких пар. Тогда ранговый коэффициент Кендалла рассчитывается по следующей формуле 
$$
\rho_{XY}^{K} = 1 - \dfrac{4Q}{m(m-1)}. 
$$
Его значения находятся в интервале от $-1$ до $+1$.

**Точечно-бисериальная корреляция**

Пусть случайная величина $X$ является ранговой, а случайная величина $Y$ является номинальной. Тогда для оценки тесноты связи можно использовать следующий коэффициент
$$
r_{XY} = \dfrac{\bar{x_1}-\bar{x_2}}{S}\sqrt{\dfrac{n_1n_2}{n(n+1)}},
$$
где $n_1$ и $n_2$ -- объемы двух групп, $n=n_1+n_2$; $\bar{x_1}, \bar{x_2}$ -- средние значения первой случайной величины в двух группах, образованных в соответствии со второй случайной величиной;
$$
S^2 = \dfrac{n_1S_1^2+n_2S_2^2}{n}, \quad S_1^2 = \dfrac{1}{n_1-1}\sum\limits_{i=1}^{n_1}(x_i-\bar{x_1})^2, \quad S_2^2 = \dfrac{1}{n_2-1}\sum\limits_{i=1}^{n_2}(x_i-\bar{x_2})^2. 
$$

В зависимости от типа шкал, в которых измерены значения случайных величин $X$ и $Y$ принято использовать следующие коэффициенты корреляции.

|Тип шкалы для $X$|Тип шкалы для $Y$|Коэффициент корреляции|
|--:|:--|:--:|
|Интервальная или отношений|Интервальная или отношений|Коэффициент Пирсона|
|Ранговая, интервальная или отношений|Ранговая, интервальная или отношений|Коэффициент Спирмена|
|Ранговая|Ранговая|Коэффициент Кендела|
|Номинальная|Номинальная|Коэффициент Гудмена-Краскела|
|Номинальная|Ранговая|Рангово бисериальный коэффициент|
|Номинальная|Интервальная или отношений|Бисериальный коэффициент|
|Интервальная|Ранговая| -- |


Важные понятия:

Значимость коэффициентов корреляции. 

Ложная корреляция. Если между двумя исследуемыми величинами установлена тесная зависимость, то из этого еще не следует их причинная  взаимообусловленность. За счет эффектов одновременного влияния неучтенных факторов смысл истинной связи может искажаться. Поэтому такую корреляцию часто называют «ложной». Для выявления «ложной» корреляции используются частные корреляции.

Частные коэффициенты корреляции. Если две переменные коррелируют, всегда можно предположить, что эта корреляция обусловлена влиянием третьей переменной, как общей причины совместной изменчивости первых двух переменных. Для проверки этого предположения достаточно третьей переменной исключить влияние этой и вычислить корреляцию двух переменных без учета влияния третьей переменой (при фиксированных ее значениях). Корреляция, вычисленная таким образом называется частной. Пусть имеются случайные величины $X$, $Y$ и $Z$. Частный коэффициент корреляции $r_{XY,Z}$ рассчитывается по следующей формуле
$$
r_{XY, Z} = \dfrac{r_{XY} - r_{XZ}r_{YZ}}{\sqrt{(1-r_{XZ}^2)(1-r_{YZ}^2)}}.
$$

В общем случае

$$
r_{XY, Z} = \dfrac{-R_{ij}}{(R_{ii}R_{jj})^{1/2}}
$$
где $R_{ij}$ -- алгебраическое дополнение для элемента $r_{ij}$ в корреляционной матрице.

Множественный коэффициент корреляции $R_{y,X}$ используется в качестве измерителя статистической тесноты связи между результирующим показателем $y$ и набором объясняющих переменных $X$
$$
R^2_{y,X} = 1 - \dfrac{\det R}{R_{00}}.
$$

## Регрессионный анализ

Пусть имеется некоторая выборочная совокупность объектов, каждый из которых описывается единым набором признаков $x_1, x_2, \ldots, x_n, y$. Представим эти данные в виде матрицы <<объект-свойство>>

$$
\begin{array}{ccccc}
x_{11} & x_{12} & \ldots  & x_{1n} & y_1\\
x_{21} & x_{22} & \ldots  & x_{2n} & y_2\\
\vdots  & \vdots & \ddots  & \vdots & \vdots\\
x_{m1} & x_{m2} & \ldots  & x_{mn} & y_m
\end{array}
$$  

В данной выборочной совокупности выделим так называемые зависимые (результирующие, целевые) и независимые (объяняющие) признаки. Например, семейный доход -- зависимый признак, заработная плата, доход от депозита, инвестиций -- независимые признаки.   

Под регрессией понимается функциональная зависимость между объясняющими переменными и условным математическим ожиданием зависимой переменной, которая, как правило, строится для предсказания значений зависимой переменной при фиксированных значениях объясняющих переменных.

```python
x = np.random.normal(1, 2, 50)
z = np.random.normal(0, 1, 50)
y = 2*x + 3 + z
f = 2*x + 3
plt.figure(figsize=(6, 6))
plt.scatter(x,y,
           color = 'green',
           edgecolors = 'black')
plt.plot(sorted(x),sorted(f))
plt.xlabel('Случайная величина X')
plt.ylabel('Случайная величина Y')
plt.title('Диаграмма рассеивания -- стастистическая зависимость')
```

К задачам регрессионного анализа относятся:
1. Определение формы зависимости между показателями.
2. Определение функции зависимости.
3. Построение прогноза.

Регрессии можно классифицировать по следующим основаниям: числу переменных, форме зависимости, характеру зависимости, типу соединения.

Относительно числа переменных различаю парную регрессию (одна объясняющая переменная) и множественную регрессию (две и более объясняющих переменных).

По форме зависимости выделяют линейную регрессию, выражаемую линейной функцией, и нелинейную, выражаемую нелинейной функцией.

В зависимости от характера регрессии различают положительную и отрицательную регрессии. Положительная регрессия имеет место, если с увеличением (уменьшением) значений объясняющей переменной значение зависимой переменной также соответственно увеличивается (уменьшается). В противном случае регрессия отрицательная.

Относительно типа соединения явлений выделяют непосредственную, косвенную и ложную регрессии. В случае непосредственной регрессии зависимая и объясняющая переменные связаны непосредственно друг с другом. Для косвенной регрессии характерно действие объясняющей переменной на зависимую через совокупность других переменных. Ложная регрессия возникает при формальном подходе к исследуемым явлениям без уяснения того, какие причины обусловливают данную связь.

К основным этапам построения уравнения регрессии относятся:
1. Постановка задачи -- формулировка цели исследования, выбор показателей для моделирования.
2. Априорный этап -- анализ сущности изучаемого процесса.
3. Информационный этап -- сбор необходимых статистических данных, наблюдаемых значений переменных.
4. Спецификация -- выбор формы уравнения регрессии для моделирования зависимости между переменными.
5. Параметризация -- оценивание параметров уравнения регрессии на основе выбранного метода.
6. Верификация -- анализ адекватности или качества построенного уравнения регрессии.

Коротко:

Регрессия -- класс задач обучения с учителем, когда по определенному набору признаков объекта нужно предсказать целевую переменную.

Задача регрессии -- нахождение зависимости между целевый и объясняющими признаками, если целевой признак выражен действительным числом. Например, определить стоимость дома по его площади.

Целевое значение -- любое действительное число.

**Линейная регрессия**

Различают:
- парную линейную регрессии;
- множественную линейную регрессии. 

**Множественная линейная регрессия**

Модель множественной линейной регрессии можно представить в следующем виде
$$
y = w_0 + w_1x_1 + w_2x_2 + \ldots + w_n x_n + \varepsilon = w_0 + \sum\limits_{i=1}^{n}w_ix_i + \varepsilon,   
$$
где $y$ -- зависимая (объясняемая) переменная или результативный признак; $x_1, x_2,\ldots, x_n$ -- независимые (объясняющие) переменные, регрессоры; $w_0, w_1, w_2, \ldots, w_n$ -- параметры, подлежащие определению; $\varepsilon$ -- ненаблюдаемая случайная составляющая.

Замечание: принято выделять следующие нелинейные формы регресии:

|Формула|Комментарий|
|--:|:--|
|$$y=w_0\cdot x_1^{w_1} \cdot x_2^{w_2} \cdot \ldots \cdot x_n^{w_n} e^\varepsilon $$|степенная|
|$$y=e^{w_0 + w_1x_1 + w_2x_2 + \ldots + w_n x_n + \varepsilon}$$|экспоненциальная|
|$$y=w_0\cdot w_1^{x_1} \cdot w_2^{x_2} \cdot \ldots \cdot w_n^{x_n} e^\varepsilon $$|показательная|
|$$y=\dfrac{1}{w_0 + x_1^{w_1} + x_2^{w_2} + \ldots + x_n^{w_n} e^\varepsilon }$$|гиперболическая|

С помощью преобразований данные формы регрессии можно привести к линейной (см. Кобзарь, стр. 649).

В основе регрессионного анализа лежат следующие предпосылки:
1. В модели $\varepsilon_i$ -- случайная, $x_1, x_2,\ldots, x_n$ -- неслучайные величины.
2. Математическое ожидание случайной ошибки $E(\varepsilon_i) = 0$.
3. Дисперсия случайной ошибки является постоянной величиной $V(\varepsilon_i) = \sigma^2$. 
4. Случайные ошибки разных наблюдений статистически независимы -- $E(\varepsilon_i, \varepsilon_j) = 0$.
5. Векторы значений объясняющих переменных линейно независимы.
6. $\varepsilon_i$ -- нормально распределенная случайная величина, $\varepsilon_i \sim N(0,\sigma^2)$.

Модель множественной регрессии, удовлетворяющая предпосылкам 1-6, называется классической нормальной линейной моделью множественной регрессии. Для построения эмпирического уравнения регрессии достаточно выполнения предпосылок 1-5. Поэтому, если не выполняется предпосылка 6, модель называется классической линейной моделью множественной регрессии. Выполнимость предпосылки 6 позволяет оценить точность построенного уравнения регрессии.

Для нахождения оценок коэффициентов модели $w_i$ ($i=0,1,2,\ldots, n$) могут использоваться различные методы, например, метод наименьших модулей, метод наименьших квадратов, метод моментов, метод максимального правдоподобия.

Согласно МНК вектор оценок $\hat{w}_0, \hat{w}_1, \hat{w}_2, \ldots, \hat{w}_n$ находится из условия минимизации остаточной суммы квадратов отклонений наблюдаемых значений зависимой переменной $y_i$ от модельных значений $\hat{y}_i$:
$$
S(\hat{w}_0, \hat{w}_1, \hat{w}_2, \ldots, \hat{w}_n) = \sum\limits_{i=1}^{n}(y_i - (\hat{w}_0 + \hat{w}_1x_1 + \hat{w}_2x_2 + \ldots + \hat{w}_n x_n) )^2 \to \min
$$

Поиск оценок $w_i$ ($i=0,1,2,\ldots, n$) сводится к решению системы линейных уравнений.

Матричный подход к описанию модели линейной множественной регрессии.

Введем следующие обозначения:

вектор значений зависимой переменной 
$$y = (y_1, y_2, \ldots, y_n)$$
вектор параметров $$w = (w_0, w_1, w_2, \ldots, w_n)$$
вектор случайных ошибок $$\varepsilon = (\varepsilon_1, \varepsilon_2, \ldots, \varepsilon_n)$$
матрица значений объясняющих переменных $$
X= \left( \begin{array}{ccccc}
1 & x_{11} & x_{12} & \ldots  & x_{1n} \\
1 & x_{21} & x_{22} & \ldots  & x_{2n} \\
\vdots  & \vdots & \ddots  & \vdots & \vdots\\
1 & x_{m1} & x_{m2} & \ldots  & x_{mn} 
\end{array} \right)
$$  

Тогда модель линейной множественной регрессии можно записать в матричной форме

$$
y = Xw + \varepsilon
$$

Условие минимизации запишется в виде 
$$
S(\hat{w}_0, \hat{w}_1, \hat{w}_2, \ldots, \hat{w}_n) = (y-X\hat{w})^{T}(y-X\hat{w}) \to \min
$$

Вектор МНК-оценок параметров модели определяется в виде
$$
\hat{w} = (X^TX)^{-1}X^Ty.
$$


Общее качество подгонки регрессии характеризует коэффициент детерминации.

Коэффициентом детерминации называется величина 
$$
R^2 = \dfrac{\sum\limits_{i=1}^{n}(\hat{y}_i-\bar{y})^2}{\sum\limits_{i=1}^{n}(y_i-\bar{y})^2}
$$

Коэффициент детерминации представляет собой меру соответствия оцененного уравнения регрессии выборочным наблюдениям и характеризует долю вариации зависимой переменной $y$, обусловленной регрессией. Коэффициент детерминации $R^2 \in [0;1]$. $R^2 =1$ тогда и только тогда, когда $\hat{y}_i = y_i$. 

Численный подход к оценке вектора параметров $$w = (w_0, w_1, w_2, \ldots, w_n)$$.

**Градиентный спуск**

Градиент -- это вектор, в направлении которого функция быстрее всего растет.

Антиградиент (вектор, противоположный градиенту) -– вектор, в направлении которого функция быстрее всего убывает.

Процедура численного поиска параметров $w_0^{(0)}, w_1^{(0)}, w_2^{(0)}$.

1. Инициализируем параметры $w_0^{(0)}, w_1^{(0)}, w_2^{(0)}, \ldots, w_n^{(0)}$. 
2. На каждом следующем шаге обновляем веса, сдвигаясь в направлении антиградиента функции потерь $S$:
$$
\begin{array}{l}
w_0^{(k)} = w_0^{(k-1)} - \nabla S(w_0^{(k-1)}), \\ 
w_1^{(k)} = w_1^{(k-1)} - \nabla S(w_1^{(k-1)}),  \\
\ldots\\
w_n^{(k)} = w_n^{(k-1)} - \nabla S(w_n^{(k-1)}).
\end{array}
$$

В матричном виде 
$$
w^{(k)} = w^{(k-1)} - \nabla S(w^{(k-1)}).
$$

Критерий остановки итерационной процедуры:
$$
\Delta w = |w^{(k)}-w^{(k-1)}|<\delta
$$

Метрики для задачи регрессии:

Среднеквадратическая ошибка (MSE):
$$
MSE\left(w_0^{(0)}, w_1^{(0)}, w_2^{(0)}\right) = \dfrac{1}{n}\sum\limits_{i=1}^{n}(\hat{y}_i-y_i)^2
$$

Среднеквадратичная ошибка (RMSE):
$$
RMSE\left(w_0^{(0)}, w_1^{(0)}, w_2^{(0)}\right) = \sqrt{\dfrac{1}{n}\sum\limits_{i=1}^{n}(\hat{y}_i-y_i)^2}
$$

Средняя абсолютная ошибка (MAE):
$$
MAE\left(w_0^{(0)}, w_1^{(0)}, w_2^{(0)}\right) = \dfrac{1}{n}\sum\limits_{i=1}^{n}|\hat{y}_i-y_i|
$$

Средняя абсолютная процентная ошибка (MAPE):
$$
MAPE\left(w_0^{(0)}, w_1^{(0)}, w_2^{(0)}\right) = \dfrac{1}{n}\sum\limits_{i=1}^{n}\dfrac{|\hat{y}_i-y_i|}{|y_i|}\times 100
$$

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# деление выборки на обучающую и тестовую
from sklearn.model_selection import train_test_split 
# задание объекта LinearRegression
from sklearn.linear_model import LinearRegression
#метрики
from sklearn.metrics import mean_squared_error,  r2_score, mean_absolute_percentage_error
# Подробно о метриках можно посмотреть здесь 
# https://scikit-learn.org/stable/modules/model_evaluation.html
```

```python
X = df[['total_area']].values
y = df['last_price'].values

X = np.array(X).reshape(-1,1)
y = np.array(y).reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)  
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(X_train, y_train)
```

```python
y_pred = linear_regression.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
