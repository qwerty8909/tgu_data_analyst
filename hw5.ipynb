```python
# Подключение библиотек
import sqlite3
import pandas as pd
```

```python
# Подключение к БД
con = sqlite3.connect('book_simple.db')
cur = con.cursor() 
```

```python
# 1. Выведите авторов, имеющихся в базе?
author_table = """
SELECT name
FROM author;
""" 
pd.read_sql(author_table,con)
```

```python
# 2. Найдите всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия
author_name_table = """
SELECT name
FROM author
WHERE name LIKE ('С. А. %');
""" 
pd.read_sql(author_name_table,con)
```

```python
# 3. Найдите все издательства, название которых состоит из двух и более слов.
publisher_table = """
SELECT name
FROM publisher
WHERE name LIKE ('% %');
""" 
pd.read_sql(publisher_table,con)
```

```python
# 4. Выведите список книг, выпущенных в издательстве "Мир".
mir_table = """
SELECT book.title
FROM book, publisher
WHERE book.publisher_id = publisher.id
AND publisher.name = 'Мир';
""" 
pd.read_sql(mir_table,con)
```

```python
# 5. Выведите количество книг, принадлежащих каждой из категорий. 
#    Результат отсортировать по возрастанию.
category_table = """
SELECT category.title,
       COUNT(*) AS total
FROM book, category
WHERE book.category_id = category.id
GROUP BY category.title
ORDER BY total;
""" 
pd.read_sql(category_table,con)
```

```python
# 6. Выведите количество книг, выпущенных в каждом издательстве. 
#    В итоговом списке оставьте только те издательства, которые выпустили 3 и более книг. 
#    Результат отсортировать по убыванию.
publisher_total_table = """
SELECT publisher.name,
       COUNT(*) AS total
FROM book, publisher
WHERE book.publisher_id = publisher.id
GROUP BY publisher.name
HAVING COUNT(*) >= 3
ORDER BY total DESC;
""" 
pd.read_sql(publisher_total_table,con)
```

```python
# 7. Выведите список книг, автором которых являлся <<С. А. Айвазян>>.
aivazyan_table = """
SELECT book.title
FROM book, authorbook, author
WHERE book.id = authorbook.book_id AND authorbook.author_id = author.id
AND author.name Like ('С. А. Айвазян%');
""" 
pd.read_sql(aivazyan_table,con)
```

```python
# 8. В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>. 
#    Результат отсортировать по убыванию.
aivazyan_publisher_table = """
SELECT publisher.name,
       COUNT(*) AS total
FROM book, authorbook, author, publisher
WHERE book.id = authorbook.book_id AND authorbook.author_id = author.id AND book.publisher_id = publisher.id
AND author.name Like ('С. А. Айвазян%')
GROUP BY publisher.name
ORDER BY total DESC;
""" 
pd.read_sql(aivazyan_publisher_table,con)
```
