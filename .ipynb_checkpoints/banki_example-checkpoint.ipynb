## 1. Общее описание

В распоряжении имеются данные о клиентах банка, расположенного в Ярославле, Ростове Великом и Рыбинске. Каждый клиент характеризуется следующим набором данных:

|Признак|Описание признака|
|---:|:---|
|`userid`|идентификатор пользователя|
|`score`|баллы кредитного скоринга|
|`City`|город|
|`Gender`|пол|
|`Age`|возраст|
|`Objects`|количество объектов в собственности|
|`Balance`|баланс на счете|
|`Products`|количество продуктов, которыми пользуется клиент|
|`CreditCard`|есть ли кредитная карта|
|`Loyalty`|активный клиент|
|`estimated_salary`|заработная плата клиента|
|`Churn`|ушел или нет|

Заказчику проекта необходимо разработать меры по **удержанию клиентов**. Для этого ему необходимо иметь представление и том, какие клиенты склонны к оттоку.

Однако в правой границе графика имеем резкий скачок -- это означает, что вдруг количество клиентов с высоким баллом стало расти. Примем этот факт также за аномалию. 

Итак, построенные графики позволяют сделать вывод о том, что выбросами будем считать записи, для одновременно выполняются следующие ограничения:
- признак $400  \geqslant{\rm score} \geqslant 840$;
- признак ${\rm age} \geqslant 62$;
- признак $42 000  \geqslant {\rm balance} \geqslant 200 000$.

$\blacktriangleright$ Сформируем датафрейм, не содержащий аномальных записей и оценим потери в данных

```python
df_clients = df[((df.score <= 840) &(df.score >= 400) & 
                 (df.age <= 62) &
                 ((df.balance <= 200000) &
                  (df.balance >= 42000)))]
```

```python
sh_old = df.shape[0]
sh_new = df_clients.shape[0]
print('Количество аномальных наблюдений {:d}, что составляет {:.2%} от общего числа записей.'
      .format(sh_old-sh_new, (sh_old-sh_new)/sh_old))
```

$\blacktriangleright$ Поскольку количество аномальных записей велико, то их следует проанализировать отдельно. Сформируем датафрейм, в котором будут содержаться только аномальные записи. Соберем эти записи через признак `user_id` и воспользуемся операциями со множествами. 

```python
user_id_anomals = list(set(df.user_id.to_list()) - set(df_clients.user_id.to_list()))
df_user_anomals = df.loc[df['user_id'].isin(user_id_anomals)]
df_user_anomals.shape[0]
```

**Промежуточный вывод:** На данном этапе провели исследование признаков на аномальные значения. Установлено, что записей с аномальными значениями признаков немногим больше 40% от объема исходной выборки. Удалять столько много данных не целесообразно. В итоге сформировали два датафрейма, содержащие:
- записи без аномальных значений признаков;
- записи с аномальными значениями признаков.

В дальнейшем оба датафрейма будут подвергнуты анализу.

### 4.1 Изучение признаков по группам: интервальные признаки

$\blacktriangleright$ Для [**интервальных признаков**](#interval_var), выделенных выше, исследуем средние по группам **для датафрейма без аномалий**. В качестве оценок средних воспользуемся `выборочном средним` и `медианой`. 

```python
df_mean = df_clients.loc[:, df_clients.columns != 'user_id'].groupby('churn')[int_var].agg('mean').T
df_median = df_clients.loc[:, df_clients.columns != 'user_id'].groupby('churn')[int_var].agg('median').T
```

```python
df_temp = pd.concat([df_mean, df_median], axis = 1)
df_temp.columns = ['Выборочное_среднее_для_оставшихся',
                   'Выборочное_среднее_для_ушедших',
                  'Медиана_для_оставшихся',
                  'Медиана_для_ушедших'] 
```

$\blacktriangleright$ Для [интервальных признаков](#interval_var), выделенных выше, исследуем средние по группам **для датафрейма с аномалиями**

```python
df_mean = df_user_anomals.loc[:, df_user_anomals.columns != 'user_id'].groupby('churn')[int_var].agg('mean').T
df_median = df_user_anomals.loc[:, df_user_anomals.columns != 'user_id'].groupby('churn')[int_var].agg('median').T
```

### 4.5 Классификация: логистическая регрессия

Изначально в целях определения степени влияния признаков на значение целевого признака `churn` планировалось построить классификационную модель с помощью логистической регрессии. Корреляционный анализ позволил в некоторой степени определить наиболее значимые признаки, но мы не будем отступать от намеченного плана.

$\blacktriangleright$ Сбалансированность классов

Перед началом решения задачи классификации исследуем распределение объектов по целевому классу. Желательно получить "примерно" равномерно распределение по классам. В противном случае в выборке имеется несбалансированность классов. 

Четкие критерии сбалансированности/несбалансированности автору неизвестны, но будем придерживаться [следующего критерия:](https://ld86.org/ml-slides/unbalanced.html#/)

`Можно считать, что выборка несбалансирована, когда размеры классов отличаются более, чем в 10 раз.`   

Надо заметить, что для задачи `прогнозирования оттока` дисбаланс классов в целом нормальное явление (как правило ушедших всегда меньше, чем оставшихся).

В случае, если выборка является несбалансированной, можно придерживаться [следующих рекомендаций](https://www.machinelearningmastery.ru/tactics-to-combat-imbalanced-classes-in-your-machine-learning-dataset/).

Необходимо заметить, что в отношении несбалансированности классов есть и [другое мнение](https://habr.com/ru/post/349078/).

```python
df_ml.groupby('churn')['churn'].agg('count')/df_ml.shape[0]*100
```

**Промежуточный вывод:** Видим, что записей, принадлежащих классу `0`, значительно больше, чем записей класса `1`, но не в 10!! раз.  

$\blacktriangleright$ Формирование обучающих и тестирующих множеств. Масштабирование признаков

Здесь
1. Выделяем набор объясняющих переменных и целевой признак;
2. Разбиваем на обучающую и тестирующие выборки в соотношении 70/30;
3. Масштабируем признаки.

```python
X = df_ml.drop(['churn'], axis = 1)
y = df_ml['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)
```

$\blacktriangleright$ Логистическая регрессия 
1. Задаем параметры для поиска на сетке;
2. Обучаем и перебираем на сетке;
3. Выводим лучшие параметры модели.

```python
grid_searcher = GridSearchCV(
    LogisticRegression(max_iter=3000, random_state=42, solver = 'liblinear'), 
    param_grid={
        'C': [0.01, 0.1, 0.5, 1, 10, 50],
        'penalty' : ['l1', 'l2']
    },
    scoring='accuracy',
    cv=5
)
```

```python
grid_searcher.fit(X_train_st, y_train);
```

```python
grid_searcher.best_params_
```

$\blacktriangleright$ Обучаем модель с оптимальными параметрами, оцениваем метрики `accuracy`, `precision` и `recall`.    

Запомним также коэффициенты при признаках в уравнении логистической регрессии.  

```python
lr = LogisticRegression(random_state=42, max_iter=3000, solver = 'liblinear', C=0.01, penalty = 'l2')
lr.fit(X_train_st, y_train)
y_pred = lr.predict(X_test_st)
lr_metrix = pd.Series([accuracy_score(y_test, y_pred), 
                       precision_score(y_test, y_pred),
                      recall_score(y_test, y_pred)],
                      ['Accuracy', 'Precision', 'Recall'])
lr_imp_var = pd.DataFrame({'feature':X_train.columns, 'coeff':lr.coef_.tolist()[0]}, )
lr_imp_var['coeff'] = lr_imp_var['coeff'].apply(abs)
lr_imp_var.sort_values(by = 'coeff', ascending = False, inplace = True)
```

Метрики качества классификационной модели

```python
lr_metrix
```

**Промежуточный вывод по качеству модели:** 2 метрики из 3 говорят о приемлемом результате классификации.

Важности признаков в модели логистической регрессии

```python
y_label = lr_imp_var['feature'].to_list()
fig, ax = plt.subplots(figsize = (15,5))
ax = sns.barplot(x='coeff', y='feature', data = lr_imp_var)
ax.get_yticklabels(y_label)
ax.set_title('Распределение важности признаков в модели логистической регрессии',fontsize = 14)
plt.xlabel('',fontsize = 12)
plt.ylabel('', fontsize = 12);
```

**Промежуточный вывод по важности коэффициентов при признаках в логрегрессии:** Все шесть признаков `age`, `loyalty`, `city_rostov`, `city_yaroslavl`, `gender_w` и `gender_m` , выделенные на этапе корреляционного анализа, как признаки, оказывающие наибольшее влияние на целевой признак, находятся в топ-6 по значимости в уравнении логрессии.  

Считаю, что результат превзошел ожидания!

### 4.6 Кластеризация: агломеративный алгоритм по Уорду и k-means

Поскольку нас интересуют, прежде всего, портреты клиентов, склонных к оттоку, то выделим из анализируемого датафрейма ушедших клиентов и выполним кластеризацию. Для полной картины кластеризуем и оставшихся клиентов.

Для выполнения кластеризации (сегментации) клиентов реализуем следующие этапы:
- Разделим выборку на ушедших и оставшихся;
- Сформируем признаковое пространство, убрав целевой признак;
- Масштабируем данные;
- Реализуем иерархическую агломеративную кластеризацию методом .linkage() и для визуализации результата методом .dendrogram() построим дендрограмму. На даннном этапе по дендрограмме, задав уровень отсечения, можно установить оптимальное число кластеров. Здесь, однако, не будем полагаться только на дендрограмму, а воспользуемся еще и методом локтя;
- Определив число кластеров, обучим модель кластеризации на основании алгоритма k-means и спрогнозируем кластеры клиентов;
- Исследуем средние значения признаков для кластеров;
- Для каждого полученного кластера посчитаем долю оттока;
- Построим распределения признаков для кластеров.

$\blacktriangleright$ Кластеризация ушедших клиентов

```python
df_ml_churn = df_ml[df_ml.churn == 1]
X = df_ml_churn.drop(['churn'], axis = 1)
scaler = StandardScaler()
X_st = scaler.fit_transform(X)
```

$\blacktriangleright$ Алгомеративная кластеризация и дендрограмма

```python
plt.figure(figsize=(15, 10))
Z = linkage(X_st, method='ward') 
plt.title('Дендрограмма для ушедших клиентов')
plt.xlabel('Объекты/ кластеры')
plt.ylabel('Расстояние')
dendrogram(Z, orientation='top')
plt.show();
```

**Промежуточный вывод:** Видим 4 класса. Далее воспользуемся еще "методом локтя".

$\blacktriangleright$ Усеченная дендрограмма: визуализация заданного числа кластеров

```python
plt.figure(figsize=(15, 10))
dendrogram(Z, 
           truncate_mode = 'lastp',
           p=4,                    
           show_contracted = True)
plt.title('Усеченная дендрограмма для ушедших клиентов')
plt.xlabel('Объекты/ кластеры')
plt.ylabel('Расстояние')
plt.show();
```

```python
agl_cluster_label = fcluster(Z, 4, criterion='maxclust')
agl_cluster_label
```

$\blacktriangleright$ Метод "локтя". Воспользуемся данным методом для определения оптимального числа кластеров

## 5. Проверка гипотез

По результатам проведенного исследования получим следующий портрет клиента, склонного к оттоку:
- возраст около 44 лет;
- является не активным;
- является женщиной;
- живет в Ростове Великом;
- пользуется одним банковским продуктом.

Необходимо заметить следующее, чем большему числу перечисленных признаков удовлетворяет клиент, тем он более склонен к оттоку.

Перейдем теперь к проверке соответствующих статистических гипотез в отношении выделенных признаков. 

Сформулируем гипотезы:
- для признака `age`: нулевая гипотеза $H_0$: Разница в возрасте между оставшимися и ушедшими клиентами статистически незначима;
- для признака `loyalty`:  нулевая гипотеза $H_0$: Для ушедших разница долей активных и неактивных клиентов статистически незначима;
- для признака `gender`: нулевая гипотеза $H_0$: Для ушедших разница долей клиентов-женщин и клиентов-мужчин статистически незначима;
- для признака `city`: нулевая гипотеза $H_0$: Для ушедших разница долей клиентов из Ростова Великого и клиентов из других городов статистически незначима; 
- для признакаи `products`: нулевая гипотеза $H_0$: Для ушедших разница долей клиентов, пользующихся ровно 1 банковским продуктов, и клиентов с 2 и более банковскими продуктами статистически незначима;

Дополнительно требуется проверить гипотезу:
- для признака `estimated_salary`: нулевая гипотеза $H_0$: Разница в заработной плате между оставшимися и ушедшими клиентами статистически незначима.  

Для всех сформулированных гипотез $H_0$ альтернативная гипотеза $H_1$ такова:
- разница статистически значима.

В целях лаконичности автор позволил себе не формулировать подробно 6 гипотез. 

Далее необходимо установить критерии для проверки гипотез. Для признаков:
- `age`, `estimated_salary` будем использовать критерий Уилкоксона-Манна-Уитни, поскольку этот критерий устойчив к аномалиям (признак `age`) и не требует нормального распределения (признак `estimated_salary`);
- `loyalty`, `gender`, `city` и `products` будем использовать $z$-критерий как для сравнения параметров биноминальных распределений.

Для всех гипотез уровень значимости установим $\alpha = 0.05$.

**ЗАМЕЧАНИЕ:** Гипотезы проверяются на данных исходного датафрейма.

 $\blacktriangleright$ Уровень значимости

```python
alpha = 0.05
```

$\blacktriangleright$ Функции, реализующие проверку гипотез 

Подготовим две функции, реализующие проверки гипотез по критерий Уилкоксона-Манна-Уитни и $z$-критерию

```python
def test_mannwhitneyu(col, alpha):
    ds1 = df[df.churn == 0][col]
    ds2 = df[df.churn == 1][col]
    results = st.mannwhitneyu(ds1, ds2, alternative = 'two-sided')
    print('Результаты проверки гипотезы H_0 по критерию Уилкоксона-Манна-Уитни:')
    print('Уровень значимости alpha={:.2f}'.format(alpha))
    print('P-value: {:.2f}'.format(results.pvalue))
    if results.pvalue < alpha:
        print(f"""Отвергаем нулевую гипотезу: разница между оставшимися и ушедшими клиентами 
по признаку <<{dict_to_rus[col]}>> статистически значима.""")
    else: print(f"""Не получилось отвергнуть нулевую гипотезу: разница между оставшимися и ушедшими клиентами 
по признаку <<{dict_to_rus[col]}>> статистически незначима.""")
    print('------------------------------------------------------------')
```

```python
def test_z_criterion(col, alpha):
    gen_group_shape = df[col].value_counts()
    goal_group_shape = df.groupby(col)['churn'].sum()
    p = goal_group_shape / gen_group_shape
    p_combined = goal_group_shape.sum() / gen_group_shape.sum()
    difference = p[0] - p[1]
    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1/gen_group_shape[0] + 1/gen_group_shape[1]))
    distr = st.norm(0, 1) 
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('Результаты проверки гипотезы H_0 по z-критерию:')
    print('Уровень значимости alpha={:.2f}'.format(alpha))
    print('P-value: {:.2f}'.format(p_value))
    if p_value < alpha:
        print(f"""Отвергаем нулевую гипотезу: для ушедших разница в долях 
по признаку <<{dict_to_rus[col]}>> статистически значима.""")
    else: print(f"""Не получилось отвергнуть нулевую гипотезу: для ушедших разница в долях по 
признаку <<{dict_to_rus[col]}>> статистически незначима.""")
    print('------------------------------------------------------------')
```

$\blacktriangleright$ Проверка гипотез для признаков `age` и `estimated_salary` 

```python
test_int_var = ['age', 'estimated_salary']
for col in test_int_var:
    test_mannwhitneyu(col, alpha)
```

Все было бы хорошо, но каждый из признакиов `city` и `products` принимают более 2-ух значений. Для них составим два вспомогательных признака, которые будут бинарные и прекрасно подходить для проверки сформулированных выше гипотез.    

Добавим два новых признака в исходный датафрейм и в словарь для удобства перевода

```python
dict_to_rus['from_rostov'] = 'Клиент из Ростова Великого'
dict_to_rus['one_product'] = 'Один банковский продукт'
```

```python
test_bin_var = ['loyalty', 'gender', 'from_rostov', 'one_product']
test_int_var = ['age', 'estimated_salary']
for col in test_bin_var:
    test_z_criterion(col, alpha)
```
